```python
# Standardize column names and check for missing values
df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
missing_summary = df.isnull().sum().sort_values(ascending=False)

# Display missing value summary
missing_summary
```

```python
df['life_expectancy'].fillna(df['life_expectancy'].median(), inplace=True)
df['adult_mortality'].fillna(df['adult_mortality'].median(), inplace=True)

# Fill Hepatitis B, Polio, and Diphtheria with the median of their groups by 'status'
df['hepatitis_b'] = df.groupby('status')['hepatitis_b'].transform(lambda x: x.fillna(x.median()))
df['polio'] = df.groupby('status')['polio'].transform(lambda x: x.fillna(x.median()))
df['diphtheria'] = df.groupby('status')['diphtheria'].transform(lambda x: x.fillna(x.median()))

# Fill alcohol and total_expenditure with their medians
df['alcohol'].fillna(df['alcohol'].median(), inplace=True)
df['total_expenditure'].fillna(df['total_expenditure'].median(), inplace=True)
```
